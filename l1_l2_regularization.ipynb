```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression # Changed 'Sklearn' to 'sklearn'
from sklearn.metrics import accuracy_score
```

```python
#load the iris dataset
iris =load_iris()
x=iris.data #petal lenght ,width etc
y=(iris.target !=0).astype(int) #binary classsifcation
```

```python
#normalization the data
scalar=StandardScaler() #keeping range
x_normalization=scalar.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test =train_test_split(x_normalization,y,test_size=0.2,random_state=42)
```

#l1 function (lesso)

```python
#regulization is in between of 0 to 1
# so when we use both l1 and l2 spicfy l1 as 0.5
model_l1=LogisticRegression(penalty='l1',solver='liblinear') #penalty amount of features is reducted in l1 by changing weight into 0
model_l1.fit(x_train,y_train) # compresss the dataset->solver
```

```python
y_pred_l1=model_l1.predict(x_test)
accuracy_l1=accuracy_score(y_test,y_pred_l1)
print(f"Accuracy with L1 regularization: {accuracy_l1*100:.2f}")
```

#l2 function (ridge)

```python
model_l2=LogisticRegression(penalty='l2',solver='liblinear')
model_l2.fit(x_train,y_train)
```

```python
y_pred_l2=model_l2.predict(x_test)
accuracy_l2=accuracy_score(y_test,y_pred_l2)
print(f"Accuracy with L2 regularization: {accuracy_l2*100:.2f}")
```

#combine things

```python
model_l1_l2=LogisticRegression(penalty='elasticnet',solver='saga',l1_ratio=0.5,) # l1_ratio -> here we are taking the average of the model
model_l1_l2.fit(x_train,y_train)
```

```python
y_pred_l1_l2=model_l1_l2.predict(x_test)
accuracy_l1_l2=accuracy_score(y_test,y_pred_l1_l2)
print(f"Accuracy with l1 and l2 regularization: {accuracy_l1_l2*100:.2f}")
```

Solvers and Regularization Compatibility
Solver	Supports L1?	Supports L2?	Works Well For
'liblinear'	✅ Yes	✅ Yes	Small datasets, binary classification


'lbfgs'	❌ No	✅ Yes	Large datasets, multiclass problems



'newton-cg'	❌ No	✅ Yes	Large datasets, multiclass problems


'sag'	❌ No	✅ Yes	Large datasets, sparse data


'saga'	✅ Yes	✅ Yes	Large datasets, supports both L1 & L2
